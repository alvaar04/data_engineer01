# Jupyter Notebook processing stage

```python
import pandas as pd
```

```python
DATA_PATH = "../data/raw_data.json"

def fetch_data() -> pd.DataFrame:
    """
    Funcion que lee los datos en formato json
    """
    data_df = pd.read_json(DATA_PATH)

    return data_df

# Cargamos los datos
data_df = fetch_data()
```
